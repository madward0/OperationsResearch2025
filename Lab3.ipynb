```python
import pandas as panda
import numpy as number
import matplotlib.pyplot as plot
from sklearn.linear_model import LinearRegression as LR

data = panda.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Sets_For_Stats/master/CuratedDataSets/hockey.csv')

x = data.loc[:, ['G']]
y = data.PTS

model = LR()
model.fit(x, y)

plot.scatter(x, y, c='blue', label='Actual Points')
plot.plot(x, model.predict(x), c='orange', label='Regression Line')
plot.xlabel('Goals Scored')
plot.ylabel('Total Points')
plot.title('Relationship Between Goals and Points')
plot.legend()
plot.show()

print(f"Regression slope: {model.coef_[0]:.2f}")
print(f"Regression intercept: {model.intercept_:.2f}")
```

### Linear Regression

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Sets_For_Stats/master/CuratedDataSets/Avengers')

df = df.dropna(subset=["Death1"])

X = df[["Appearances", "Year", "Gender", "Current?"]].copy()
y = df["Death1"]

le_gender = LabelEncoder()
X["Gender"] = le_gender.fit_transform(X["Gender"].astype(str))

le_current = LabelEncoder()
X["Current?"] = le_current.fit_transform(X["Current?"].astype(str))

model = LogisticRegression(max_iter=1000)
model.fit(X, y)

thor_row = df[df["Name/Alias"].str.contains("Thor Odinson", case=False, na=False)]

if not thor_row.empty:
    thor_X = thor_row[["Appearances", "Year", "Gender", "Current?"]].copy()
    thor_X["Gender"] = le_gender.transform(thor_X["Gender"].astype(str))
    thor_X["Current?"] = le_current.transform(thor_X["Current?"].astype(str))
    
    prob = model.predict_proba(thor_X)[0,1]
    print("Probability Thor Odinson dies (Death1=1):", prob)
else:
    print("Thor Odinson not found in dataset")
```
