```python
import random

def find_short_bottle():
    pulls = 0
    while True:
        pulls += 1
        bottle = 10 if random.random() < 0.9 else 9
        if bottle == 9:
            return pulls
trials = 1000
results = [find_short_bottle() for _ in range(trials)]
print(f"Average pulls needed: {sum(results)/len(results):.2f}")
```

```python
import random

bottles = [10]*9 + [9]
random.shuffle(bottles)

def find_short_realistic():
    pulls = 0
    for bottle in bottles:      
        pulls += 1
        if bottle == 9:
            return pulls
trials = 1000
results = [find_short_realistic() for _ in range(trials)]
print(f"Average pulls needed: {sum(results)/len(results):.2f}")
```

```python
import random

plates = ['pink']*6 + ['blue']*6 + ['black']*6 + ['green']*6
trials = 1000

count_single = 0
for _ in range(trials):
    if random.choice(plates) == 'pink':
        count_single += 1
count_four = 0
for _ in range(trials):
    temp = plates.copy()
    random.shuffle(temp)
    draw = [temp.pop() for _ in range(4)]
    if 'pink' in draw:
        count_four += 1

print("P(single pink) ≈", count_single/trials)
print("P(at least one pink) ≈", count_four/trials)
```

```python
import random

plates = ['pink']*6 + ['blue']*6 + ['black']*6 + ['green']*6
trials = 1000
exact_one = 0
at_least_one = 0

for _ in range(trials):
    stack = plates.copy()
    dirty = [stack.pop() for _ in range(12)]
    random.shuffle(dirty)
    stack.extend(dirty)
    draw = [stack.pop() for _ in range(4)]
    pinks = draw.count('pink')
    if pinks == 1: exact_one += 1
    if pinks >= 1: at_least_one += 1

print("P(exactly one pink) ≈", exact_one/trials)
print("P(at least one pink) ≈", at_least_one/trials)
```

```python
import random

def pull_with_rejection(stack):
    rejected = []
    while True:
        plate = stack.pop()
        if plate == 'pink':
            stack = rejected + [plate] + stack
            return stack
        else:
            rejected.append(plate)

plates = ['pink']*6 + ['blue']*6 + ['black']*6 + ['green']*6
stack = plates.copy()
random.shuffle(stack)

for _ in range(1000):
    stack = pull_with_rejection(stack)

print("Top 10 plates after rejection:", stack[:10])
```
